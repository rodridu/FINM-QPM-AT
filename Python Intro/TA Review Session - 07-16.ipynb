## Message Type

Use the ***class*** keyword to define a class

__init__ is a constructor function used to define member variables at initiation

Use the double underscore to define private class member functions

```python
class Message:
    def __init__(self, sending_time, sequence_number):
        self.__sending_time = sending_time
        self.__sequence_number = sequence_number
        
    def getSendingTime(self):
        return self.__sending_time
        
    def setSendingTime(self, sending_time):
        self.__sending_time = sending_time
        
    def getSequenceNumber(self):
        return self.__sequence_number
        
    def setSequenceNumber(self, sequence_number):
        self.__sequence_number = sequence_number
```

#### Inheritance: Defining Child Classes from Parent Classes

```python
class AddModifyOrderMessage(Message):
    def __init__(self, sending_time, sequence_number, price, quantity, side, order_id):
        super().__init__(sending_time, sequence_number)
        self.__price = price
        self.__quantity = quantity
        self.__side = side
        self.__order_id = order_id
```
