```python
# Please complete the calculate_sentiment function below
# It takes in the dataframe "df_sa" with polarity scores
# Please achieve the following:
# 1. Calculate the mean of the "compound" column and round it to 2 decimal place for each ticker, and we treat this value as the mean sentiment of that stock
# 2. Create a new dataframe with 2 columns: 'Ticker' and 'Mean Sentiment'
# 3. Sort the new dataframe by mean sentiment in decending order
# 4. Return the sorted dataframe
# HINT: You might want to take a look at the given dataframe "df_sa" first

def calculate_sentiment(df_sa):
    # You need to get a list of the tickers contained in the provided dataframe
    unique_ticker = df_sa.Ticker.unique()# -- -- #
    # Creat an empty list to store the mean value
    values = []
    # Iterate each ticker and calculate the mean sentiment
    for ticker in unique_ticker: 
        # Filter the rows containing the ticker
        dataframe = df_sa[df_sa['Ticker']==ticker]# -- -- #
        # Set the new index as "Ticker"
        dataframe.set_index('Ticker')  # -- -- #
        # Remove the "Headline" column
        dataframe.drop('Headline', axis=1) # -- -- #
        # Calculate the mean of the "compound" column and round it to 2 decimal places
        mean = dataframe.compound.mean().round(2)# -- -- #
        # Append the mean to the list "values"
        values.append(mean)
    
    # Create a new dataframe that maps the ticker and its mean sentiment
    df = pd.DataFrame({'Ticker' : unique_ticker,'Mean Sentiment' : values})# -- -- # 
    # Set the index as "Ticker"
    df.set_index('Ticker',inplace=True)  # -- -- #
    # Sort the dataframe by mean sentiment in decending order
    df.sort_values(by=['Mean Sentiment'],ascending=False,inplace=True) # -- -- #
    
    return df
```

```python
def histogram_weekly_losses(results):
    # create histogram for the weekly losses
    # use the folloing functions with bins=50
    pnl_diff = results.Pnl.diff(5)
    weekly_losses = pnl_diff[pnl_diff<0]
    weekly_losses,_=np.histogram(weekly_losses,bins=50)
    return weekly_losses


def histogram_monthly_losses(results):
    # create histogram for the monthly losses
    # use the folloing functions with bins=50
    pnl_diff = results.Pnl.diff(20)
    monthly_losses = pnl_diff[pnl_diff<0]
    monthly_losses,_=np.histogram(monthly_losses,bins=50)
    return monthly_losses
    

def max_draw_down(results):
    #calculate max drawdown
    drawdown_max_pnl = np.max(results['Pnl'])
    return round(drawdown_max_pnl,2)


def histogram_position_holding_times(results):
    #calculate histogram for the position holding times
    nrows = len(results)
    position_holding_times = []
    cur_holding_times = 0
    for i in range(0,nrows):
        if results.Position[i] != 0:
            cur_holding_times += 1
        elif cur_holding_times != 0:
            position_holding_times.append(cur_holding_times)
            cur_holding_times = 0
    
    histogram_position_holding_times,ht=np.histogram(position_holding_times,bins=50)
    pdf=pd.DataFrame(histogram_position_holding_times,
                     index=ht[:-1],columns=["count"])
    pdf.index.name="bins"
    return pdf

def volatility_summary(results):
    from statistics import stdev
    weekly_pnls=results.Pnl-results.Pnl.shift(5)
    weekly_pnls=weekly_pnls[0::5]
    #from statistics import stdev
    weekly_pnls=weekly_pnls.dropna()
    weekly_losses=weekly_pnls[weekly_pnls<0]
    sharpe_ratio=weekly_pnls.mean()/stdev(weekly_pnls)
    sortino_ratio=weekly_pnls.mean()/stdev(weekly_losses)
    # give the volatlity summary for:
    print('PnL Standard Deviation:', stdev(weekly_pnls))
    print('Sharpe ratio:', sharpe_ratio)
    print('Sortino ratio:', sortino_ratio)
    pdf=pd.DataFrame([stdev(weekly_pnls),sharpe_ratio,sortino_ratio],
                     index=['PnL Standard Deviation','Sharpe ratio:','Sortino ratio:'],columns=['Summary'])
    return pdf


def traded_volume_summary(results):
    # calculate the total traded volume
    traded_volume = np.abs(results['Trades']).sum()*10 # you need to use the abs function on the correct field
    print('Total traded volume:', traded_volume)
    return traded_volume
```
